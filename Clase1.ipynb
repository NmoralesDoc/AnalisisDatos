# **2. Limpieza de Datos**

La **limpieza de datos** es una actividad esencial dentro de los proyectos de Machine Learning. Es posible que los datos sin pre-procesamiento tengan errores voluntarios e involuntarios que pueden afectar a la precisión de los modelos de ML y de esta manera proveer resultados incorrectos. Es importante entonces llevar adelante tareas de limpieza de datos.


# **2.1 ¿Qué es la Limpieza de Datos?**

La limpieza de datos es un proceso esencial en Machine Learning y en Data Science. Engloba varios procesos destinados a mejor la calidad de los datos. Para esto se disponen de herramientas y actividades o buenas prácticas a seguir.


Estos procesos se usan para **corregir y eliminar registros incorrectos** en un dataset. Esto significa identificar y sustituir o eliminar los datos o registros incompletos, incorrectos, corruptos o poco relevantes.

**La limpieza de datos es diferente de la transformación de datos.** La limpieza convierte los datos de un formato a otro, mientras que la transformación (o wrangling) convierte los datos en un formato adecuado para el análisis.

# **2.2 ¿Para qué sirve la Limpieza de Datos?**

En la Era de la Información, los datos son un recurso esencial para las empresas y organismos. Y los necesitamos constantemente para la Toma de Decisión.

Según un estudio de IBM, la mala calidad de los datos cuesta tan solo en USA unos 3100M USD/año. Por otra parte, Accenture informó que la calidad de los datos es el principal escollo del uso de IA en Salud. En términos de proyectos, una campaña de Marketing con datos de baja calidad, puede estar mal orientada y fracasar; en Salud, puede llevar a tratamientos o diagnósticos inadecuados; en Supply Chain, puede ocasionar problemas de inventario y entregas, entre otros.

La prevención a través de la Limpieza de Datos tiene un costo bajo. Arreglar los problemas existentes se estima en un costo 10 veces mayor. Y arreglar un problema en los datos después de un fallo, 100 veces más.

Algo que no es menor: la limpieza de datos es necesaria para cumplir con las normas de privacidad como el [RGPD](https://es.wikipedia.org/wiki/Reglamento_General_de_Protecci%C3%B3n_de_Datos).

# **2.3 Ventajas de la Limpieza de Datos**

A priori, se eliminan errores en las Bases de Datos.

Más allá de eso, disponer de datos de calidad genera un efecto positivo en todas las actividades relacionadas con datos, y 
se aplica a todas las áreas del conocimiento como así también en el ámbito empresarial y en organismos públicos.

Lograr esto (datos de alta calidad), es posible a través de buenas prácticas en todas las áreas y de manera continua.

# **2.4 Pasos de la Limpieza de Datos**

En empresas y organismos, la limpieza de datos debe ser un proceso incremental establecido en un Plan de Calidad de Datos.

Principalmente, se deben identificar la fuente de errores y problemas y especificar las acciones a realizar. Estas acciones deben ser llevadas adelante por los responsables de las áreas correspondientes.

Es importante medir la calidad de los datos. Entre otras cosas, esto permite priorizar y focalizar acciones específicas.

Las acciones especificadas ante los diferentes tipos de errores, con el tiempo se actualizarán según el monitoreo de la calidad de los datos y la actividad de la organización.

Además de los errores ocurridos internamente, es importante detectar y corregir los datos en origen. Esto significa actuar antes que ingresen a los diferentes sistemas informáticos.

Por otra parte, también es importante monitorear a los datos consumidos en tiempo real.

En específico las acciones a realizar son:

1.   **Identificar datos que deben ser limpiados**.
Esto implica:

  * Eliminar observaciones no deseadas (aquellas que pueden ser correctas pero no son importantes para el Proyecto)
  * Eliminar observaciones irrelevantes
  * Eliminar observaciones duplicadas


2.   **Corregir errores estructurales**. Entre ellos aparecen errores de nomenclatura, errores tipográficos y similares. Su no tratamiento puede generar clases incorrectas o mal etiquetadas.


3.   **Detectar outliers y corregir missing values**.
Los outliers son valores atípicos. Su origen puede ser un error de carga de datos, almacenamiento o parseo. Incluso, puede ser un valor cierto. Es importante detectarlos para considerar su inclusión o no como datos en los modelos y de ser así, un tratamiento adecuado.

  Los valores ausentes también deben detectados. Ya sea para su descarte o bien para su reemplazo por valores acordes.

Técnicas de tratamiento de valores faltantes

Media

```python
media_edad = round(df['age'].mean()) #Obtenemos la media de la columna edad 
media_edad 
```

```python
df['age'].fillna(media_edad, inplace=True) #Reemplazamos los valores nulos de la columna edad por la media de la edad 
```

```python
media_fare = round(df['fare'].mean()) #Obtenemos la media de la columna fare 

media_fare 

df['fare'].fillna(media_fare, inplace=True) #Reemplazamos los valores nulos de la columna fare por la media de la fare 
```

```python
#moda
df['boat'].fillna(df['boat'].mode()[0], inplace=True)
```

```python
print(df.isnull().sum(axis = 0)) #Chequeamos que ya no existan valores nulos en la columna edad
```
